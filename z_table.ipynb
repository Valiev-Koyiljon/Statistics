#z_table using python and its pandas library

```python
# necessary libraries for our small project
import math
import pandas as pd


# range of z-scores to be included in the z-table
z_scores = [round(i * 0.1, 1) for i in range(0, 40)]


# standard normal cumulative distribution function
def phi(x):
    return (1.0 + math.erf(x / math.sqrt(2.0))) / 2.0


# corresponding probabilities using the standard normal distribution
probabilities = [round(phi(z), 4) for z in z_scores]


# DataFrame to store the z-table
z_table = pd.DataFrame({
    'z': z_scores,
    '0.00': probabilities,
    '0.01': probabilities[1:] + [0],
    '0.02': probabilities[2:] + [0, 0],
    '0.03': probabilities[3:] + [0, 0, 0],
    '0.04': probabilities[4:] + [0, 0, 0, 0],
    '0.05': probabilities[5:] + [0, 0, 0, 0, 0],
    '0.06': probabilities[6:] + [0, 0, 0, 0, 0, 0],
    '0.07': probabilities[7:] + [0, 0, 0, 0, 0, 0, 0],
    '0.08': probabilities[8:] + [0, 0, 0, 0, 0, 0, 0, 0],
    '0.09': probabilities[9:] + [0, 0, 0, 0, 0, 0, 0, 0, 0]
})


# Setting the z-score column as the index
z_table.set_index('z', inplace=True)
z_table
```

```python
# converting dataframe to excel file
"""It will be saved in your project folder in xlsx format. you can open this using MS excel."""
z_table.to_excel("z_table.xlsx")
```
